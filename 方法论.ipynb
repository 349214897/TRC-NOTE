# 怎么组织一次对话或文章
- 学术方面
>1. 对于学术性的对话和文章一般可以遵循下面顺序（类似于毕业论文）
    - 首先问题的产生背景和解决问题的意义
    - 问题的准确定义
    - 问题用什么方法
    - 这个方法的步骤是什么，每一步进行下去的理由
    - 循环往复，不断把问题按照这种方法，由大到小切分，最终能够有条理的把一个问题讲清楚  
    
# 工作和学习的关系
- 工作是对学习的巩固，因为学习到的东西不经高强度的重复训练就不能很好的固话到自己的大脑，不能稳固学到的东西，做到信手拈来，缺少了工作上的内容需求，则学习犹如空中楼阁，无的放矢
- 学习是对工作的提高，没有了学习的理论指导和多新知识的了解，对系统的整体梳理与掌握，那么工作就一直停留在重复造轮子的局面，不会有境界的提升和水平的提升  
**综上，工作学习是一个巩固提升的关系**

# 怎么迅速适应一份新的工作
- 首先，观察是最重要的，在刚进入时，必须迅速摸清每个人的性格特点和能力，摸清楚整个团队的氛围和一些约定俗成的东西，团队里每个人之间的分工，职级，以及对应的他们之间的人际关系，自己的工作职责，定位，整个组的技术方案，工作内容，用到的软件工具，工作的评价指标
- 再者，分清哪些是对自己考核升迁评价至关重要的，哪些是自己可以依赖的盟友，哪些是和自己有竞争关系的，哪些是水平高的，哪些是可以去询问请教的
- 对于常用到的工具的使用技能一定要熟练，比如说excel，ppt，word的使用，在哪里都用的到，更简洁的提升逼格的类似于前面的ppt的mou（一款markdown制作生成简洁文本ppt的工具），影响笔记（富文本的网络笔记，随时记，随时分享），这些提升办公效率的工具的使用是需要在任何情景下提前熟悉的
- 对比是最能够发现问题找到差距的方法，要多了解其他人的想关信息，不断和自己对比，找出优点和不足。所谓知己知彼，百战不殆。

# 找工作面试时的原则
- 总原则，别人找你肯定是缺这方面的人，所以你只要在这一块展示自己的闪光点，让人明白进去是有你的价值的就可以
- 所以在面试过程中，一定要你主导对话，尽量往自己熟悉擅长的地方靠，主动让他发现你的闪光点，而不是去被动的接受解决面试官提出的问题，一旦他不断的抛出问题，企图挖掘你的闪光点时，你就危险了

# 深度学习新进展和人学习的大致规律
- 深度学习已有简单的框架进化到了综合的学习策略阶段，如目前的gan网络，对偶学习策略网络,增强学习网络，这是新的发展方向，另外人脑有不断丰富数据地图的功能，而人脑接受最大量的数据是图数据，所以可以利用图和大脑的结构做一点文章。
- 人的学习规律遵循了
    * 首先是事物内在规律的指导，这种规律是前辈们无数实践而获得的经验
    * 用这种规律指导自己学习，高强度的践行，从而使自己在一定时间内获得质的飞越
    * 最后集大成，把规律都忘了，只留下那个映入日常的不需经过大脑的习惯，从而达到超级效率，然后在此基础上自己探索出新的东西
- 总之就是站在巨人的肩膀上，借力发展，巩固提高，融会贯通，内化为自己的东西

# 增强行动力的方法
